#### 경기->경기도, 경남->경상남도, 부산특별시->부산광역시

```python
addr['시도']=addr['시도'].apply(lambda v:addr_aliases.get(v,v))
addr['시도'].unique()
```

#### 보건의료기관수 df, 인구수 df 병합

```python
all_data=pd.merge(addr_group,population,how='inner',left_index=True,
right_index=True)
all_data.head()
```

#### 병합한 df에서 필요없는 컬럼(시도_y,구군_y,남자인구수,여자인구수) 삭제

```python
last_data=all_data[['시도_x','구군_x','count','총인구수']]
last_data = last_data.rename(columns={'시도_x':'시도','구군_x':'구군','총인구수':'인구수'})
last_data
```

```python
mc_count=last_data['count']
mc_count
```

```python
# 인구 10만명당 공공보건기관 비율 계산
last_data['mc_ratio']=mc_count.div(last_data['인구수'],axis=0) * 1000000
last_data.head()
```
