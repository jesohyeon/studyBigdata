### 패키지 import

```python
import pandas as pd   # 파일을 읽어 데이터프레임에 저장하고 작업한 데이터를 데이터프레임 형태로 구성하여 파일에 저장하는 작업 수행 모듈
import glob   # 경로와 이름을 지정하여 파일을 다루는 파일 처리 작업을 위한 모듈
from functools import reduce   # 2차원 리스트를 1차원 리스트로 차원을 줄이기 위한 모듈
from nltk.tokenize import word_tokenize   # 자연어 처리 패키지 중 단어 토큰화 작업을 위한 모듈
from nltk.corpus import stopwords   # 자연어 처리 패키지 중 불용어 정보를 제공하는 모듈
from nltk.stem import WordNetLemmatizer   # 자연어 처리 패키지 중 단어 형태의 일반화를 위해 표제어 추출을 제공하는 모듈
from collections import Counter   # 데이터 집합에서 개수를 자동으로 계산하기 위한 모듈
import matplotlib.pyplot as plt   # 히스토그램을 그리기 위한 맵플러리 패키지의 내부 모듈
from wordcloud import STOPWORDS, WordCloud  # 워드 클라우드를 그리기 위해 사용할 워드 클라우드용 불용어 모듈과 워드 클라우드 모듈
import re   # 메타 문자를 이용하여 특정 규칙을 작성하는 정규식을 사용하기 위한 모듈
```

#### 데이터전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
all_data_concat = all_data_concat.drop('번호', axis=1) 
```

```python
all_data_concat = all_data_concat.drop(all_data_concat.columns[[0]],
axis='columns')
all_data_concat
```

```python
# 전처리 작업을 위해 nltk.corpus에서 제공하는 영어 불용어를 불러와서 저장해준다. => stopwords.words('english')
stopWords = set(stopwords.words('english'))
stopWords
```

```python
# 그리고 표제어 추출 작업을 제공하는 WordNetLemmatizer 객체를 생성한다.
lemma = WordNetLemmatizer()
lemma
```

```python
# all_title의 제목에 대해 정규식으로 만든 규칙을 적용하여 영어가 아닌것, 즉 알파벳으로 시작하지 않는 단어 '[^a-zA-Z]+'는 공백으로 치환하여 're.sub' 제거하고,
# 소문자로 정규화하고 'lower()', 단어 토큰화 'word_tokenize()'를 한다. 
# 그리고 불용어 'stopWords'를 제거한 후에 표제어 추출'lemmatize(w)'을 한다.

words = []

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+', ' ', str(title))  # 단어가 아닌것들은 공백으로 치환
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]

    words.append(EnWordsTokenStopLemma)
```

```python
print(words) # 불용어 제거 남은 키워드 단어만 남음
```

```python
# 2차원리스트 -> 1차원리스트로
words_reduce = list(reduce(lambda x, y: x+y, words))
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
# 출현횟수가 많은 상위 30개 단어'count.most_common(30) 중에서 단어 길이가 1보다 큰 것만 word_count 딕셔너리에 저장한 후 출력

word_count = dict()

for tag, counts in count.most_common(30):
    if len(str(tag)) > 1:
        word_count[tag] =counts
        # print(f'{tag} : {counts}')

word_count
```

### 단어개수 차트

```python
# 히스토그램 크기, x축 레이블과 y축 레이블을 지정
# 상위 50개만 저장한 word_count 딕셔너리에서 x축 값으로 사용할 sort_Keys와 y축 값으로 사용할 sorted_Values를 역순으로 정렬'reverse=True'하여 준비
# x축 눈금'plt.xticks'은 sorted_Keys 리스트의 값을 순서대로 사용
# 설정사항을 적용하여 히스토그램을 그림

sorted_Keys =sorted(word_count, key=word_count.get, reverse=True)
sorted_Values =sorted(word_count.values(), reverse=True)

plt.figure(figsize=(20,5))
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation=90)
plt.show()
```
