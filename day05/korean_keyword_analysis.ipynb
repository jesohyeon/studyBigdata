# 한국어 분석

## 패키지 import

```python
import json
import re
from konlpy.tag import Okt
from collections import Counter

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
```

```python
# 크롤링 데이터 로드
data = json.loads(open('./data/4차 산업혁명.json', mode='r', encoding='utf-8').read())
data
```

```python
# 데이터 메세지 중 품사가 명사인 단어만 추출
# data['message']
message=' '
for item in data:
    message = message + re.sub(r'[^\w]', ' ', item['message']) + ' ' # 단어가 아닌것들은 공백으로 치환

message
```

```python
# 명사만 추출
nlp = Okt()
message_noun = nlp.nouns(message)
message_noun
```

```python
len(message_noun) # 463개의 단어
```

### 추출한 단어 검색

```python
count = Counter(message_noun)
count
```

```python
word_count = dict()

for tag, counts in count.most_common(30):
    if len(str(tag)) > 1:
        word_count[tag]=counts
        print(f'{tag} : {counts}')
```

#### 한글 폰트 설정

```python
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font_name)
```

### 키워드 빈도수 바차트

```python
plt.figure(figsize=(12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)

plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)),list(sorted_Keys), rotation=90)

plt.show()
```

#### 한글 워드클라우드

```python
wc = WordCloud(font_path=font_path, background_color='#f5f5f5', width=1280, height=760)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(40,15))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
